```python
import random # standard library
random.seed(2024) # so pseudo-random numbers are reproducible
NUMBER_THROWS = 100_000
NUMBER_OF_DICE = 6
throw_results = []
for _ in range(NUMBER_THROWS):
    result = 0
    for dice in range(NUMBER_OF_DICE):
        result += random.randint(1, 6) # again randint is inclusive - rare for Python
    # at end of NUMBER_OF_DICE throws, append result to throw_results
    throw_results.append(result)    
# print first 5 results
print(f"First 5 results: {throw_results[:5]}")
# let's get some statistics on our results
print(f"Minimum: {min(throw_results)}")
print(f"Average: {sum(throw_results) / NUMBER_THROWS}") # could use len(throw_results) instead of NUMBER_THROWS
print(f"Maximum: {max(throw_results)}")
```

## Law of large numbers

As number of trials increases, the probability of the event will converge to the expected value.

Wiki: https://en.wikipedia.org/wiki/Law_of_large_numbers

```python
# let's get a counter of our results
from collections import Counter
counter = Counter(throw_results)
print(f"Counter: {counter}")
```

```python
# let's plot our counter as a bar char
import matplotlib.pyplot as plt
plt.bar(counter.keys(), counter.values())
plt.show()
```

```python
# now let's make a function that will throw the dice for us, count the results and plot them
# the parameters will be the number of throws and number of dice per throw
def throw_dice(throws, dice, visualize=True, save=False):
    results = []
    for _ in range(throws):
        result = 0
        for _ in range(dice):
            result += random.randint(1, 6)
        results.append(result)
    counter = Counter(results)
    if visualize:
        plt.bar(counter.keys(), counter.values())
        if save:
            plt.savefig(f"dice_throw_{dice}.png")
        plt.show()
    # return results # could return results if you want to do more

# let's test our function
throw_dice(1_000, 4, save=True)
```

```python
# let's plot from 1 to 10 dice and with 100_000 throws each
for dice in range(1, 11):
    throw_dice(100_000, dice)
```
