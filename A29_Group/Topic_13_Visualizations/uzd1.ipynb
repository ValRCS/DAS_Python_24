```python
# Uztaisiet 100 000 garu sarakstu ar 6 metam kauliņu summu. (6 metienu summa vienā mērijumā) 
# Saskaitiet šo summu biežumu (izmantojam Counter)
# Attēlojam stabiņu( vai kādā citā) diagrammā.
 
 
 
import random
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt
 
 
dice_sums = []
i = 0
counter = {}
NUM_DICE = 6
 
while i < 100000:
    r_sum = 0
    for _ in range(NUM_DICE): # _ signifies that we don't care about the value
        r_sum += random.randint(1, 6)
    dice_sums.append(r_sum)
    i += 1
 
# for sums in dice_sum:
#     counter[sums] = counter.get(sums, 0) + 1
 
# unique_dice_sum = []
# unique_dice_repeats = []
# for sum in counter:
#     repeats=counter[sum]
#     unique_dice_sum.append(sum)
#     unique_dice_repeats.append(repeats)
# xpoints = np.array(unique_dice_sum)
# ypoints = np.array(unique_dice_repeats)

# we can use Counter to count the number of times each sum appears
counter = Counter(dice_sums)
plt.title("Dice sums in chart ")
plt.xlabel("Dice sums")
plt.ylabel("Dice sums repeats")
# plt.bar(xpoints,ypoints)
plt.bar(counter.keys(), counter.values())
plt.show()
```

```python
# lets make a function to return a list of dice sums
# parameters will be number of dice and number of rolls
def roll_dice(num_dice, num_rolls):
    dice_sum = []
    for _ in range(num_rolls):
        r_sum = 0 # represents the current roll sum
        for _ in range(num_dice): # turns out we can reuse the same variable name
            r_sum += random.randint(1, 6)
        dice_sum.append(r_sum)
    return dice_sum

# test on 8 throws of 2 dice
dice_sums = roll_dice(2, 8)
print(dice_sums)
```

```python
# now let's make a function that will plot the dice sums
# parameters will be number of dice and number of rolls
def plot_dice_sums(num_dice, num_rolls):
    dice_sums = roll_dice(num_dice, num_rolls)
    counter = Counter(dice_sums)
    plt.title("Dice sums in chart ")
    plt.xlabel("Dice sums")
    plt.ylabel("Dice sums repeats")
    plt.bar(counter.keys(), counter.values())
    plt.show()
    return dice_sums

# now we let's plot dice from 1 to 8 on 10_000 rolls
# for i in range(1, 9):
#     plot_dice_sums(i, 10_000)
```

```python
# law of large numbers
# wikipedia link: https://en.wikipedia.org/wiki/Law_of_large_numbers
# now let's plot 6 dice from 1_000_000 rolls
dice_sums = plot_dice_sums(6, 1_000_000)
```

```python
# now let's create a json file with the dice sums starting from key 6 and ending at key 36
import json
# let's create a counter again
counter = Counter(dice_sums)
# let's create a dictionary with sorted keys
# one way to do it would be to sort items by key and then create a dictionary again
sorted_counter = dict(sorted(counter.items()))
# now let's save it as json file
with open("dice_sums.json", "w") as f:
    json.dump(sorted_counter, f, indent=4)
```
